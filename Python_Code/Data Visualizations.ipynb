```python
# Histogram of Review Ratings
plt.figure(figsize=(12, 6))
sns.histplot(shopping_trends["Review Rating"], kde=False, bins=30)
plt.title("Distribution of Review Ratings")
plt.show()
```

```python
# Histogram of frequency of purchases
plt.figure(figsize=(12, 6))
sns.histplot(shopping_trends["Frequency of Purchases"], kde=False, bins=30)
plt.title("Distribution of Frequency of Purchases")
plt.show()
```

```python
# Boxplot of category vs purchase amount
sns.boxplot(x='Category', y='Purchase Amount (USD)', palette='husl', data=shopping_trends)
plt.show()
```

```python
# Boxplot of shipping type vs purchase amount
sns.boxplot(x='Shipping Type', y='Purchase Amount (USD)', palette='husl', data=shopping_trends)
plt.show()
```

```python
# Boxplot of Season vs Purchase Amount
sns.boxplot(x='Season', y='Purchase Amount (USD)', palette='husl', data=shopping_trends)
plt.show()
```

```python
# Violin plot of  gender vs purchase amount
plt.figure(figsize=(10, 6))
sns.violinplot(x="Gender", y="Purchase Amount (USD)", data=shopping_trends)
plt.title("Gender vs Purchase Amount")
plt.show()
```

```python
# Count plot of categories 
plt.figure(figsize=(10, 6))
sns.countplot(x="Category", data=shopping_trends)
plt.title("Count of Purchases across different Categories")
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart of categories
category_counts = shopping_trends['Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Categories')
plt.show()
```

```python
# Density plot of purchase amounts with the hue being category
g = sns.FacetGrid(shopping_trends, hue="Category", palette='husl', height=6)
g.map(sns.kdeplot, 'Purchase Amount (USD)', shade=True).add_legend()
```

```python
# Count plot of Subscription status
plt.figure(figsize=(10, 6))
sns.countplot(x="Subscription Status", data=shopping_trends)
plt.title("Count of Subscription Status")
plt.show()
```

```python
# Scatterplot of age vs frequency of purchases with hue being gender
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age", y="Frequency of Purchases", hue="Gender", data=shopping_trends)
plt.title("Age vs Frequency of Purchases")
plt.show()
```

```python
# Count plot of payment method with a hue of frequency of purchases
plt.figure(figsize=(10, 6))
sns.countplot(x="Payment Method", hue='Frequency of Purchases', data=shopping_trends)
plt.title("Count of transactions per Payment Method")
plt.xticks(rotation=90)
plt.show()
```
