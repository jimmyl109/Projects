```python
# Checking the VIF inflation scores, we see some variables higher than the thresh hold of 10
# we also scale and set each of the variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
import pandas as pd

X = body_fat.drop('BodyFat', axis=1) # BodyFat or whatever is your dependent variable
Y = body_fat[['BodyFat']].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

vif = pd.DataFrame()
vif["variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

print(vif)
```

```python
# splitting the data into the necessary sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size = 0.3, random_state = 0)
```

```python
# I used the models that multi collinearity had no effect on as well as some models that did.
# I fitted each of the models.
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm


ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train.ravel())
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train.ravel())
elasticnet = ElasticNet(alpha=0.1, l1_ratio=0.5)
elasticnet.fit(X_train, y_train.ravel())
LR = LinearRegression()
LR.fit(X_train, y_train.ravel())
rf = RandomForestRegressor(n_estimators=20) 
rf.fit(X_train, y_train.ravel())
SVR = svm.SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
SVR.fit(X_train, y_train.ravel())
```

```python
# I then predicted the values using the rest of the testing set
y_pred_ridge = ridge.predict(X_test)
y_pred_lasso = lasso.predict(X_test)
y_pred_elasticnet = elasticnet.predict(X_test)
y_pred_LR = LR.predict(X_test)
y_pred_RF = rf.predict(X_test)
y_pred_SVR = SVR.predict(X_test)
```

```python
# combined the datasets back so it can be used for the SVR
X_combined = np.concatenate((X_train, X_test))
y_combined = np.concatenate((y_train, y_test))
```

```python
# Calculated the R^2 values of each of the models, the models had very high R^2 values so I was skeptical
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, cross_val_score
r2 = r2_score(y_test, y_pred_ridge)
print(f'Ridge R2: {r2}')
r2 = r2_score(y_test, y_pred_lasso)
print(f'Lasso R2: {r2}')
r2 = r2_score(y_test, y_pred_elasticnet)
print(f'Elasticnet R2: {r2}')
r2 = r2_score(y_test, y_pred_LR)
print(f'LR R2: {r2}')
r2 = r2_score(y_test, y_pred_RF)
print(f'RF R2: {r2}')
```

```python
# I then used 5 fold cross validation to find the average MAE of the models 
# Lasso regression and RF had the lowest MAE
k = 5
kf = KFold(n_splits=k)

skf_ridge = cross_val_score(ridge, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_absolute_error')
skf_lasso = cross_val_score(lasso, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_absolute_error')
skf_elasticnet = cross_val_score(elasticnet, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_absolute_error')
skf_LR = cross_val_score(LR, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_absolute_error')
skf_RF = cross_val_score(rf, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_absolute_error')
skf_SVR = cross_val_score(SVR, X_combined, y_combined.ravel(), cv=5, scoring='neg_mean_absolute_error')

print(f' Avg MAE:{-skf_ridge.mean()}')
print(f' Avg MAE:{-skf_lasso.mean()}')
print(f' Avg MAE:{-skf_elasticnet.mean()}')
print(f' Avg MAE:{-skf_LR.mean()}')
print(f' Avg MAE:{-skf_RF.mean()}')
print(f' Avg MAE:{-skf_SVR.mean()}')
```

```python
## I then used 5 fold cross validation to find the average MSE of the models 
# Most of the models had similar MSE
from sklearn.model_selection import KFold, cross_val_score

k = 5
kf = KFold(n_splits=k)

skf_ridge = cross_val_score(ridge, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_lasso = cross_val_score(lasso, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_elasticnet = cross_val_score(elasticnet, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_LR = cross_val_score(LR, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_RF = cross_val_score(rf, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_SVR = cross_val_score(SVR, X_combined, y_combined.ravel(), cv=5, scoring='neg_mean_squared_error')

print(f' Avg MSE:{-skf_ridge.mean()}')
print(f' Avg MSE:{-skf_lasso.mean()}')
print(f' Avg MSE:{-skf_elasticnet.mean()}')
print(f' Avg MSE:{-skf_LR.mean()}')
print(f' Avg MSE:{-skf_RF.mean()}')
print(f' Avg MSE:{-skf_SVR.mean()}')
```

```python
# I then used 5 fold cross validation to find the average RMSE of the models 
# Most of the models had similar RMSE
from sklearn.model_selection import KFold, cross_val_score
from math import sqrt 
k = 5
kf = KFold(n_splits=k)

skf_ridge = cross_val_score(ridge, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_lasso = cross_val_score(lasso, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_elasticnet = cross_val_score(elasticnet, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_LR = cross_val_score(LR, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_RF = cross_val_score(rf, X_train, y_train.ravel(), cv = kf, scoring='neg_mean_squared_error')
skf_SVR = cross_val_score(SVR, X_combined, y_combined.ravel(), cv=5, scoring='neg_mean_squared_error')

print(f' Avg RMSE:{sqrt(-skf_ridge.mean())}')
print(f' Avg RMSE:{sqrt(-skf_lasso.mean())}')
print(f' Avg RMSE:{sqrt(-skf_elasticnet.mean())}')
print(f' Avg RMSE:{sqrt(-skf_LR.mean())}')
print(f' Avg RMSE:{sqrt(-skf_RF.mean())}')
print(f' Avg RMSE:{sqrt(-skf_SVR.mean())}')
```
